```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing

```

```python
heart_data = pd.read_csv('/content/heart.csv')
```

```python
heart_data
```

```python
heart_data.head()
```

```python
x= heart_data.drop('target', axis=1)
y= heart_data['target']
```

```python
print('x=' ,x)
print('y=' ,y)
```

```python
heart_data.isnull().sum()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test, y_pred) * 100
print("Accuracy:", accuracy)
```

```python
model_gaus = GaussianNB()
model_gaus.fit(x_train, y_train)
```

```python
y_pred_gaus = model_gaus.predict(x_test)
```

```python
accuracy_gaus = accuracy_score(y_test, y_pred_gaus) * 100
print("Accuracy:", accuracy_gaus)
```

```python
from sklearn.naive_bayes import BernoulliNB
model_ber = BernoulliNB(fit_prior= False)
model_ber.fit(x_train, y_train)
```

```python
ypred_ber= model_ber.predict(x_test)
```

```python
accuracy_ber = accuracy_score(y_test, ypred_ber) * 100
print("Accuracy:", accuracy_ber)
```

```python
model_ber_true = BernoulliNB(fit_prior= True)
model_ber_true.fit(x_train, y_train)
```

```python
ypred_ber_true= model_ber_true.predict(x_test)
```

```python
accuracy_ber_true = accuracy_score(y_test, ypred_ber_true) * 100
print("Accuracy:", accuracy_ber_true)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_tree = DecisionTreeClassifier(splitter='random')
model_tree.fit(x_train, y_train)
```

```python
y_pred_tree = model_tree.predict(x_test)
```

```python
accuracy_tree = accuracy_score(y_test, y_pred_tree) * 100
print("Accuracy:", accuracy_tree)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=5)
model_knn.fit(x_train, y_train)
```

```python
y_pred_knn = model_knn.predict(x_test)
```

```python
accuracy_knn = accuracy_score(y_test, y_pred_knn) * 100
print("Accuracy:", accuracy_knn)
```

```python
precision_gaus = precision_score(y_test, y_pred_gaus) *100
recall_gaus = recall_score(y_test, y_pred_gaus) *100
f1_gaus = f1_score(y_test, y_pred_gaus) *100
print(f'The metrics for Gaussian NB classifier are:  precision = {precision_gaus}, recall = {recall_gaus}, f1_score = {f1_gaus}')
```

```python
precision_ber = precision_score(y_test, ypred_ber) *100
recall_ber = recall_score(y_test, ypred_ber) *100
f1_ber = f1_score(y_test, ypred_ber) *100
print(f'The metrics for Bernoulli NB with fit prior set to false classifier are:  precision = {precision_ber}, recall = {recall_ber}, f1_score = {f1_ber}')
```

```python
precision_ber_true = precision_score(y_test, ypred_ber_true) *100
recall_ber_true = recall_score(y_test, ypred_ber_true) *100
f1_ber_true = f1_score(y_test, ypred_ber_true) *100
print(f'The metrics for Bernoulli NB with fit prior set to true classifier are: precision = {precision_ber_true}, recall = {recall_ber_true}, f1_score = {f1_ber_true}')
```

```python
precision_tree = precision_score(y_test, y_pred_tree) *100
recall_tree = recall_score(y_test, y_pred_tree) *100
f1_tree = f1_score(y_test, y_pred_tree) *100
print(f'The metrics for Decision Tree classifier are: precision = {precision_tree}, recall = {recall_tree}, f1_score = {f1_tree}')
```

```python
precision_knn = precision_score(y_test, y_pred_knn) *100
recall_knn = recall_score(y_test, y_pred_knn) *100
f1_knn = f1_score(y_test, y_pred_knn) *100
print(f'The metrics for KNN classifier are: precision = {precision_knn}, recall = {recall_knn}, f1_score = {f1_knn}')
```
